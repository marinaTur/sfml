```python
y = train['TARGET']
```

```python
train = train.drop('TARGET', axis=1)
```

```python
# соеденим train и test
```

```python
all_data = pd.concat([train, test])
```

```python
# необходимо удалить индекс, 
# у нас есть 0-строка в train и 0-строка в test (и так далее 1,2,3,4)
# если мы просто объеденим - у нас сохранятся индексы
# мы же просто уберем индекс и автоматически создастся новый индекс, сквозной и уникальный
all_data = all_data.reset_index(drop=True)
```

```python
# посмотрим типы данных и их заполняемость
```

```python
all_data.info()
```

```python
# Видим, что часть данных - object, скорее всего стоки.
# Давайте выведем эти значения для каждого столбца
```

```python
for i in all_data.columns: # перебираем все столбцы
    if str(all_data[i].dtype) == 'object': # если тип столбца - object
        print(i) # выводим название столбца
        print(set(all_data[i])) # выводим все его значения (но делаем set - чтоб значения не повторялись)
        print() # выводим пустую строку
```

```python
# можно заметить что некоторые переменные, 
# которые обозначены как строки (например PERSONAL_INCOME) на самом деле числа,
# но по какой-то (ниже увидим) причине были распознаны как строки
```

```python
all_data['PERSONAL_INCOME'] = all_data['PERSONAL_INCOME'].map(lambda x: x.replace(',', '.')).astype('float')
```

```python
# причина же что использовалась запятая для разделения не целой части числа
```

```python
all_data['CREDIT'] = all_data['CREDIT'].map(lambda x: x.replace(',', '.')).astype('float')
all_data['FST_PAYMENT'] = all_data['FST_PAYMENT'].map(lambda x: x.replace(',', '.')).astype('float')
all_data['LOAN_AVG_DLQ_AMT'] = all_data['LOAN_AVG_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
all_data['LOAN_MAX_DLQ_AMT'] = all_data['LOAN_MAX_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
```

```python
# запускаем повторно код и смотрим, какие столбцы с какими строковыми значениями остались
```

```python
for i in all_data.columns:
    if str(all_data[i].dtype) == 'object':
        print(i)
        print(set(all_data[i]))
        print()
```

# Данные готовы, теперь вам предлагается начать решать задачу

```python
# Есть ли пропуски в данных? что с ними сделать?
# Ваш код :-)
all_data.info()
```

```python
all_data=all_data[~all_data['WORK_TIME'].isnull()]
all_data=all_data[~all_data['PREVIOUS_CARD_NUM_UTILIZED'].isnull()]
```

```python
all_data.shape
```

```python
# Как поступить со строковыми переменными?
# Ваш код :-)
all_data = pd.get_dummies(all_data)
```

```python
# В итоге должны получиться данные без пропусков и только числовые
```

```python
all_data.info()
```

```python
# Вновь разделите на train и test (не мешая)
# Ваш код :-)
```

```python
train2=all_data[all_data['AGREEMENT_RK'].isin(train['AGREEMENT_RK'])]
```

```python
test2=all_data[all_data['AGREEMENT_RK'].isin(test['AGREEMENT_RK'])]
```

```python
print ('all_data:',all_data.shape)
print('train2:',train2.shape)
print('test2:',test2.shape)
```

```python
# Сделать hold-out sampling на train выборке
from sklearn.cross_validation import train_test_split
# test_size=0.3, random_state=42

# Ваш код :-)
y2 = y[y.index.isin(list(train2.index))]
x_train, x_test, y_train, y_test = train_test_split(train2, y2, test_size=0.3, random_state=42)
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Ваш код :-)
```

```python
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(x_train, y_train)
dtc_predict = dtc.predict(x_test)
dtc_predict_proba = dtc.predict_proba(x_test)
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
rfc_predict = rfc.predict(x_test)
rfc_predict_proba = rfc.predict_proba(x_test)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_predict = lr.predict(x_test)
lr_predict_proba = lr.predict_proba(x_test)
```

```python
rfc_predict
```

```python
lr_predict_proba[:,1]
```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall

# Ваш код :-)

from sklearn.metrics import accuracy_score,precision_score, recall_score
```

```python
print('DTC:accuracy', accuracy_score(y_test, dtc_predict))
print('DTC:precision', precision_score(y_test, dtc_predict))
print('DTC:recall', recall_score(y_test, dtc_predict))
print()
print('RFC:accuracy', accuracy_score(y_test, rfc_predict))
print('RFC:precision', precision_score(y_test, rfc_predict))
print('RFC:recall', recall_score(y_test, rfc_predict))
print()
print('LR:accuracy', accuracy_score(y_test, lr_predict))
print('LR:precision', precision_score(y_test, lr_predict))
print('LR:recall', recall_score(y_test, lr_predict))
```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально посмотреть)

# Ваш код :-)
from sklearn.metrics import precision_recall_curve

dtc_precision, dtc_recall, dtc_thresholds = precision_recall_curve(y_test, dtc_predict_proba[:,1])
rfc_precision, rfc_recall, rfc_thresholds = precision_recall_curve(y_test, rfc_predict_proba[:,1])
lr_precision, lr_recall, lr_thresholds = precision_recall_curve(y_test, lr_predict_proba[:,1])

%matplotlib inline
from matplotlib import pyplot as plt
plt.figure(figsize=(5, 5))

plt.plot(dtc_precision, dtc_recall)
plt.plot(rfc_precision, rfc_recall)
plt.plot(lr_precision, lr_recall)

# plt.plot([0, 1], [0, 1])
plt.ylabel('recall')
plt.xlabel('precision')
plt.grid(True)
plt.title('PR curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе

# Ваш код :-)

%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve

dtc_fpr, dtc_tpr, dtc_thresholds = roc_curve(y_test, dtc_predict_proba[:,1])
rfc_fpr, rfc_tpr, rfc_thresholds = roc_curve(y_test, rfc_predict_proba[:,1])
lr_fpr, lr_tpr, lr_thresholds = roc_curve(y_test, lr_predict_proba[:,1])

plt.figure(figsize=(5, 5))

plt.plot(dtc_fpr, dtc_tpr)
plt.plot(rfc_fpr, rfc_tpr)
plt.plot(lr_fpr, lr_tpr)

# plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
print ("DTC_roc_auc_score:",roc_auc_score(y_test, dtc_predict_proba[:,1]))
print ("RFC_roc_auc_score:",roc_auc_score(y_test, rfc_predict_proba[:,1]))
print ("LR_roc_auc_score:",roc_auc_score(y_test, lr_predict_proba[:,1]))
```

```python
from sklearn.cross_validation import cross_val_score
# Сделать k-fold (10 фолдов) кросс-валидацию каждой модели
# И посчитать средний roc_auc

# Ваш код :-)
```

```python

dtc_k_folds = cross_val_score(
                            dtc,
                            train2,
                            y2,
                            scoring='roc_auc',
                            cv=10
                        )

rfc_k_folds = cross_val_score(
                            rfc,
                            train2,
                            y2,
                            scoring='roc_auc',
                            cv=10
                        )

lr_k_folds = cross_val_score(
                            lr,
                            train2,
                            y2,
                            scoring='roc_auc',
                            cv=10
                        )

print ("DTC_roc_auc_score_avg:",dtc_k_folds.mean())
print ("RFC_roc_auc_score_avg:",rfc_k_folds.mean())
print ("LR_roc_auc_score_avg:",lr_k_folds.mean())
```
