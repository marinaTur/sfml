```python
import math
```

```python
def predict (features, coef, last_coef):
    s = 0
    for x1, x2 in zip(features, coef):
        s += x1*x2
    s += last_coef    
    return(
        1/(
            1 + math.e**(
                -1 * s 
            )
        )
    )
```

```python
#task_3

#Нужно взять sklearn.linear_model.LogisticRegression 
#Визуализировать как меняется ROC_AUC от параметров penalty и C
```

```python
# penalty : str, 'l1' or 'l2', default: 'l2'
#     Used to specify the norm used in the penalization. The 'newton-cg',
#     'sag' and 'lbfgs' solvers support only l2 penalties.
    
# C : float, default: 1.0
#     Inverse of regularization strength; must be a positive float.
#     Like in support vector machines, smaller values specify stronger
#     regularization.    
```

```python
penalty_lst = ['l1','l2']
#C_lst = [0.01,1.0,10.0,50.0,100.0]

plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve by penalty')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))

for n in penalty_lst:
    lr = LogisticRegression(penalty = n)
    lr.fit(X_train, y_train)
    lr_predict = lr.predict_proba(X_test)
    
    dtc_fpr, dtc_tpr, dtc_thresholds = roc_curve(y_test, lr_predict[:,1])
    plt.plot(roc_curve(y_test, lr_predict[:,1])[0], roc_curve(y_test, lr_predict[:,1])[1])
    
#penalty l1 дает повышение точности, но считается дольше    
```

```python
C_lst = [0.01,1.0,10.0,50.0,100.0]

plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve with penalty = l2 by C value')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))

for n in C_lst:
    lr = LogisticRegression(penalty = 'l2', C = n)
    lr.fit(X_train, y_train)
    lr_predict = lr.predict_proba(X_test)
    
    dtc_fpr, dtc_tpr, dtc_thresholds = roc_curve(y_test, lr_predict[:,1])
    plt.plot(roc_curve(y_test, lr_predict[:,1])[0], roc_curve(y_test, lr_predict[:,1])[1])

# C не меняет кривую?    
```

```python
C_lst = [0.01,10.0,50.0,100.0]

plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve with penalty = l1 by C value')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))

for n in C_lst:
    lr = LogisticRegression(penalty = 'l1', C = n, n_jobs=3)
    lr.fit(X_train, y_train)
    lr_predict = lr.predict_proba(X_test)
    
    dtc_fpr, dtc_tpr, dtc_thresholds = roc_curve(y_test, lr_predict[:,1])
    plt.plot(roc_curve(y_test, lr_predict[:,1])[0], roc_curve(y_test, lr_predict[:,1])[1])
```
